# Power Law Transformation

```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
img=cv.imread('img/airview.tif')
img=cv.resize(img,dsize=(0,0),fx=0.25,fy=0.25)

def gamma(f,gamma=1.0):
    f1=f/255.0
    return np.uint8(255*(f1 ** gamma))

gc=np.hstack((
    gamma(img,0.5),
    gamma(img,0.75),
    gamma(img,1.0),
    gamma(img,2.0),
    gamma(img,3.0),
    gamma(img,4.0),
    gamma(img,5.0),
    gamma(img,10.0),
    gamma(img,30.0),
    gamma(img,100.0),    
    ))
cv.imshow("gamma correction", gc)

cv.waitKey()
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(20, 10)) 
plt.imshow(cv.cvtColor(gc, cv.COLOR_BGR2RGB))
plt.show() 
```

$f'(j, i) = (L-1) \times \hat{f}(j, i)^\gamma$
